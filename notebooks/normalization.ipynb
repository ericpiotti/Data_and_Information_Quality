```python
import pandas as pd
```

```python
df = pd.read_csv('../Dataset/Comune-di-Milano-Pubblici-esercizi(in)-2.csv',sep=';',encoding='utf-8')
```

```python
# select only text columns
text_cols = df.select_dtypes(include="object").columns

# convert to lowercase
df[text_cols] = df[text_cols].apply(lambda col: col.str.lower())
df
```

```python
df = df.rename(columns={"þÿTipo esercizio storico pe": "Tipo esercizio storico pe"})
df = df.rename(columns={"Ubicazione": "Indirizzo"})
df = df.rename(columns={"Descrizione via": "Nome via"})
df
```

```python
chars = ["ÿ", "ý"]

pattern = "|".join(map(lambda x: "\\" + x, chars))

mask = df.select_dtypes(include="object").apply(
    lambda col: col.astype(str).str.contains(pattern, na=False)
).any(axis=1)

df_bad = df[mask]
df_bad
```

```python
pattern = r"\bcaff[ÿý]"

mask = df.select_dtypes(include="object").apply(
    lambda col: col.astype(str).str.contains(pattern, na=False)
).any(axis=1)

df_non_ascii = df[mask]
df_non_ascii
```

```python
count = 0
for c in df.select_dtypes(include="object").columns:
    count += df[c].astype(str).str.count(r"\bcaff[ÿý]").sum()

count
```
