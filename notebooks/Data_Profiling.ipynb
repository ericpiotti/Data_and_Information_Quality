# 2. DATA PROFILING
Data profiling is introduced as a separate step of the project. The goal of this notebook is to explore the Milan dataset and derive basic metadata from it.

The libraries used for profiling are loaded. Pandas is used to read and manipulate the tabular data, seaborn is used to produce a heatmap of correlations, and matplotlib is used to create scatter plots and activate inline plotting in the notebook environment.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

The dataset of public establishments in the city of Milan is loaded from the CSV file. The separator is set to semicolon, since fields in this file are separated by ;. The resulting DataFrame is stored in a variable called MILANO that will be used in all the following analyses

```python
MILANO = pd.read_csv("Comune-di-Milano-Pubblici-esercizi(in)-2.csv", sep=";")
MILANO
```

A first basic profiling step is to look at the list of attributes and at the overall size of the dataset. By inspecting the list of column names and the shape of the table, it is possible to understand how many variables are available and how many records will be analysed in the rest of the notebook.

```python
MILANO.columns
MILANO.shape
```

A small sample of the first tuples is displayed. This preview makes it easier to recognise what each attribute actually contains and to connect the column names to meaningful pieces of information about the public establishments.

```python
MILANO.head()
```

The data types of the attributes are examined using the dtypes property. This shows which columns are interpreted as numeric and which columns are treated as categorical or textual. This distinction will be used later when selecting subsets of numerical and categorical variables for specific analyses.

```python
MILANO.dtypes
```

The numerical and categorical attributes are explicitly separated using the select_dtypes function. The list called NUM collects the names of the numeric columns, while the list called CAT collects the names of the boolean or object columns. These lists will make it easier to run histograms and correlation analysis only on appropriate variables.

```python
NUM = list(MILANO.select_dtypes(include=['int64','float64']).columns)
NUM

CAT = list(MILANO.select_dtypes(include=['bool','object']).columns)
CAT
```

For single column analysis and cardinality measures, one attribute is chosen as an example. In this notebook the attribute Superficie somministrazione is used, since it is numerical and describes an important characteristic of each establishment. The following cells will replicate the computations of length, count, distinct values and the related quality measures.

```python
MILANO["Superficie somministrazione"]
```

The total number of rows is computed using the len function on the chosen column. The number of rows and the number of columns are also obtained from the shape of the DataFrame, and their product gives the total number of cells in the table. These quantities match the definitions used in the exercise and provide a first idea of the size of the dataset.

```python
ROWS = len(MILANO["Superficie somministrazione"])
ROWS

MILANO.shape[0]
MILANO.shape[1]

CELLS = MILANO.shape[0] * MILANO.shape[1]
CELLS
```

The number of actual (non-null) values for the chosen attribute is measured using the count function. Then the value_counts function is applied to the same column to see how many times each distinct value appears.

```python
COUNT = MILANO["Superficie somministrazione"].count()
COUNT

MILANO["Superficie somministrazione"].value_counts()
```

The cardinality of the chosen attribute is computed using the nunique function, which counts the number of distinct values actually present. This cardinality represents how many different surfaces appear in the dataset. The result is stored in a variable called DISTINCT.

```python
DISTINCT = MILANO["Superficie somministrazione"].nunique()
DISTINCT
```

Uniqueness is evaluated following the definition from the profiling exercise. The ratio between the cardinality and the total number of rows is computed. This value expresses how often the surface value uniquely identifies a record in the dataset. A value close to one means that almost all values are different, while a lower value means that many rows share the same surface.

```python
UNIQUENESS = DISTINCT / ROWS
print("UNIQUENESS:", UNIQUENESS)
```

Distinctness is computed as the ratio between the cardinality and the number of actual values. This measure ignores missing values and focuses only on the records where the attribute is present. Distinctness therefore shows how many different values exist among the non-null entries.

```python
DISTINCTNESS = DISTINCT / COUNT
print("DISTINCTNESS:", DISTINCTNESS)
```

Constancy is evaluated by finding the maximum frequency in the value counts of the attribute and dividing it by the number of actual values. First the maximum of the value_counts is computed, then this maximum count is divided by COUNT. A high constancy indicates that one particular value is very frequent compared to the others, which can reveal the presence of standard or default values.


```python
MAX_COUNTS = max(MILANO["Superficie somministrazione"].value_counts())
MAX_COUNTS

CONSTANCY = MAX_COUNTS / COUNT
print("CONSTANCY:", CONSTANCY)
```

The minimum, maximum and mode of the chosen numerical attribute are computed using the corresponding pandas functions.

```python
print("MIN:", MILANO["Superficie somministrazione"].min())
print("MAX:", MILANO["Superficie somministrazione"].max())
print("MODE:", MILANO["Superficie somministrazione"].mode()[0])
```

The extremes are extended to all numerical attributes at once. By selecting the numeric columns with the NUM list, the minimum and maximum values are computed in a vectorised way. This shows, for each numerical attribute in the dataset, the smallest and largest values that appear in the data.

```python
print("MIN:\n\n",MILANO[NUM].min())
print("\n\n")
print("MAX:\n\n",MILANO[NUM].max())
```

The average and the standard deviation of the chosen attribute are computed. These two statistics give a sense of the central tendency and dispersion of the serving surface.

```python
print("Average:", MILANO["Superficie somministrazione"].mean())
print("Standard Deviation:", MILANO["Superficie somministrazione"].std())
```

A histogram of the chosen attribute is generated. The distribution of the serving surface is plotted using the hist function. This graphical view makes it easier to detect skewness, outliers or concentration of values around particular ranges.

```python
MILANO["Superficie somministrazione"].dropna().hist()
```

Descriptive statistics are computed for all numerical attributes using the describe function. This provides, in a single table, the count, mean, standard deviation, minimum, quartiles and maximum for each numeric column. 

```python
MILANO.describe()
```

Multiple histograms are displayed at once for two numerical attributes. For the Milan dataset the attributes Codice via and Superficie somministrazione are chosen as an example. Plotting them side by side helps to compare their distributions.

```python
MILANO[["Codice via", "Superficie somministrazione"]].dropna().hist()
```

Correlation between numerical attributes is analysed using the corr function. The Pearson, Kendall and Spearman correlation coefficients are computed on the numeric subset identified by NUM. These matrices show how strongly the numerical attributes are linearly or monotonically related.

```python
MILANO[NUM].corr(method="pearson")

MILANO[NUM].corr(method="kendall")

corr = MILANO[NUM].corr(method="spearman")
corr
```

The correlation matrix computed with the Spearman method is visualised as a heatmap. The seaborn heatmap function is used to highlight stronger and weaker correlations through colour intensity. 

```python
sns.heatmap(corr)
```

As a final profiling step, a scatter plot is created between the zone code ZD and the serving surface. The plot makes it possible to visually inspect whether certain zones tend to host establishments with larger or smaller surfaces.

```python
plt.figure(figsize=(8, 6))
plt.scatter(MILANO["ZD"], MILANO["Superficie somministrazione"], marker='o', color='r', label='Data')
plt.xlabel("ZD")
plt.ylabel("Superficie somministrazione")
plt.title("SCATTER PLOT: ZD vs Superficie")
plt.legend()
plt.grid(False)
plt.show()
```

### Exploration of the numeric column *Superficie somministrazione* (and other numeric features)

The column **`Superficie somministrazione`** has 6904 rows, but only 6825 non-null values, so there are a few missing surfaces that will need to be handled later. The values are all positive and range from **2 m² to 2336 m²**, with **mode = 60 m²**, **mean ≈ 86 m²** and **standard deviation ≈ 90 m²**. This already suggests a very spread distribution with many small/medium venues and a few very large ones.

From the counts we see **400 distinct surface values**.  
- **Uniqueness** = distinct values / total rows ≈ **0.058**  
- **Distinctness** = distinct values / non-null values ≈ **0.059**  
Both ratios are low, meaning the same surface values are reused many times (e.g. 60, 59, 58, 30 m², …).  
- **Constancy** = max frequency / non-null ≈ **0.023**: the most frequent value (60 m², 155 occurrences) represents only around 2.3% of the non-missing records. So the distribution is *not* dominated by a single value; instead, it is quite granular with many different sizes.

The histogram of `Superficie somministrazione` confirms a **strong right skew**: the vast majority of establishments have surfaces between roughly 20 and 150 m², while a small number of outliers reaches several hundred or even more than 2000 m². The descriptive table (`MILANO.describe()`) is consistent with this: the median is 64 m², the 25th percentile 42 m² and the 75th percentile 100 m², so half of the venues lie in a relatively narrow band, and the large standard deviation is mainly caused by a few very big places.

When we plot **`Codice via` vs `Superficie somministrazione`**, the two histograms behave differently: `Codice via` looks almost like a uniform/flat distribution across its range, which is what we expect from an arbitrary identifier code, while `Superficie somministrazione` is clearly concentrated on lower values with a long tail on the right. This is another indication that `Codice via` should be treated as an ID, not as a meaningful numeric variable.

The correlation matrices (Pearson, Kendall, Spearman) show that:
- `Codice via` and `ZD` have a **moderate positive correlation** (~0.43), which makes sense because streets are typically located in specific zones.
- `Superficie somministrazione` has **very weak correlations** with both `Codice via` and `ZD` (coefficients close to 0, slightly negative).

The heatmap and the scatter plot **`ZD` vs `Superficie somministrazione`** confirm this: each zone contains both small and large surfaces, and there is no clear trend of bigger or smaller venues in a specific zone. The points are widely scattered for every value of `ZD`, with a few extreme outliers in different zones.

In summary, the profiling of the numeric part of the dataset tells us that:
- `Superficie somministrazione` is a valid positive numeric feature with a **heavily right-skewed distribution**, many different values and a few extreme outliers.
- There are **some missing values** in this column but not a huge amount.
- `Codice via` behaves like a numeric identifier, not like a variable with geometric meaning.
- Zone (`ZD`) and surface have **no strong relationship**, so zone alone does not explain the size of the establishments.
