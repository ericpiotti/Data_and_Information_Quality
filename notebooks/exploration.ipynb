```python
nulls = {}
for col in ESERCIZI.columns:
    nulls[col] = ESERCIZI[col].isnull().sum()
    
print(nulls)
```

```python
# Number of not null values. Clearly there are some many null values in 'Insegna'
ESERCIZI.count()
```

```python
# Perform the completeness evaluation:
total_values = ESERCIZI.shape[0] * ESERCIZI.shape[1]
total_values
```

```python
not_nulls = ESERCIZI.count().sum()
not_nulls
```

```python
completeness = not_nulls / total_values
print(f"{completeness*100:.2f}%") 
# 89.51%
```

```python
# Dive deeper on the 'Insegna' column. it is 50.61%!!! Extremely low!!
rows = ESERCIZI.shape[0]
non_nulls_insegna = ESERCIZI['Insegna'].count()
completeness_insegna = non_nulls_insegna / rows
print(f"{completeness_insegna*100:.2f}%") 
```

```python
# Compute the completeness without the column 'Insegna'
dataset_without_insegna = ESERCIZI.drop(columns=['Insegna'])
total_values_without_insegna = dataset_without_insegna.shape[0]*dataset_without_insegna.shape[1]
non_nulls_without_insegna = dataset_without_insegna.count().sum()
completeness_without_insegna = non_nulls_without_insegna / total_values_without_insegna
print(f"{completeness_without_insegna*100:.2f}%")
```
